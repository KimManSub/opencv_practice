### 2-1. 보간법 사용하기

```python
import cv2
from google.colab.patches import cv2_imshow

# 이미지 읽기
image = cv2.imread('like_lenna.png')

# 이미지 크기 변경 (다양한 보간법 사용)

image_nearest = cv2.resize(image, (224, 224), interpolation=cv2.INTER_NEAREST)
# 가장 빠르고, 정확한 보간을 원하지 않는 경우 사용

image_linear = cv2.resize(image, (224, 224), interpolation=cv2.INTER_LINEAR)
# 기본 보간법으로, 바르면서도 비교적 좋은 품질 제공

image_cubic = cv2.resize(image, (224, 224), interpolation=cv2.INTER_CUBIC)
#이미지 품질을 중시할 때 사용하되, 속도가 느려질수 있음

image_lanczos = cv2.resize(image, (224, 224), interpolation=cv2.INTER_LANCZOS4)
#고품질이 필요할 때 사용하지만, 속도가 더 느릴수 있음

# 각 보간법 적용된 이미지 출력
cv2_imshow(image_nearest)
cv2_imshow(image_linear)
cv2_imshow(image_cubic)
cv2_imshow(image_lanczos)
```

```python
# 배율로 사이즈 변환

image_big = cv2.resize(image, None, fx=2, fy=2, interpolation=cv2.INTER_LINEAR)
cv2_imshow(image_big)
```

```python
#대칭 변환

image_flip = cv2.flip(image, 0)
cv2_imshow(image_flip)

image_flip2 = cv2.flip(image, 1)
cv2_imshow(image_flip2)

image_flip2 = cv2.flip(image, -1)
cv2_imshow(image_flip2)

# 0  : 수직 방향 뒤집기
# 1  : 수평 방향 뒤집기
# -1 : 수평 및 수직 으로 모두 뒤집기
```

```python
#회전변환

height, width = image.shape
matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)
#회전의 중심점 width/2, height/2 기준으로  반시계방향 90도 회전 ,1(비율은 그대로-> 만약 2라면 2배 아니면 0.5
result = cv2.warpAffine(image, matrix, (width, height))

cv2_imshow(result)
```

```python
#회전변환2

matrix = cv2.getRotationMatrix2D((width/2,height/2),30,1)
result = cv2.warpAffine(image,matrix,(width,height),borderValue=200)
#원본이미지,출력 이미지크기 설정,빈영역의 색상을 지정(기본)
cv2_imshow(result)
```

### 3. 도형 그리기

```python
# 선 그리기 : cv2.line
# 원 그리기 : cv2.circle
# 직사각형 그리기 cv2.rectangle
# 타원 그리기 : cv2.ellipse
# 다각형 그리기 : cv2.polylines, cv2.fillpoly
```

```python


space = np.zeros((500, 1000), dtype=np.uint8)
#크기 500,1000, 색상 검은색, 데이터 타입 0~255인
line_color = 255
#선 색은 255 흰색 0은 검정
space = cv2.line(space, (100, 100), (800, 400), line_color, 3, 1)
# 왼쪽 0,0이 기본임 오른쪽으로 갈수록 x 값 증가 아래쪽으로 갈수록 y값 증가
# 선픽셀 3픽셀
# cv2.LINE_4: 4-connected 선.(채택)
# cv2.LINE_8 (기본값): 8-connected 선.
# cv2.LINE_AA: 안티에일리어싱 처리된 부드러운 선.
cv2_imshow(space)
```

```python
space = np.zeros((500, 1000), dtype=np.uint8)
color = 255
space = cv2.circle(space, (600, 200), 100, color, 4, 1)
#원의 중심 600,200 ,반지름 100 ,선색 255,선굵기 4픽셀, 1
#1 : cv2.LINE_AA (안티에일리어싱).
#4 : cv2.LINE_4.
#8 : cv2.LINE_8.
cv2_imshow(space)
```

```python
space = np.zeros((768, 1388), dtype=np.uint8)
color = 255
space = cv2.rectangle(space, (500, 200), (800, 400), color, 5, 1)
#왼쪽위,오른족 아래 꼭지점 좌표를 기준으로 사각형을 그림
cv2_imshow(space)
```

```python
space = np.zeros((768, 1388), dtype=np.uint8)
color = 255
space = cv2.ellipse(space, (500, 300), (300, 200), 0, 90, 250, color, 4)
#500,300 위치에 중심을 두고 가로 300, 세로 200 인 타원을 그리며,시작 각도 90도에서 250까지 그려지는 타원을 그림=

cv2_imshow(space)
```

```python
space = np.zeros((768, 1388), dtype=np.uint8)
color = 255
obj1 = np.array([[300, 500], [500, 500], [400, 600], [200, 600]])
# 4점으로 이루어진 사각형
obj2 = np.array([[600, 500], [800, 500], [700, 200]])
# 3점으로 이루어진 삼각형
space = cv2.polylines(space, [obj1], True, color, 3)
#그릴이미지,다각형 점들 지정한 배열,닫힌형태로 그림, 선색깔=255, 선두께
space = cv2.fillPoly(space, [obj2], color, 1)
#그릴이미지,다각형 점들 지정한 배열, 다각형 내부를 채울 색, 다각형 채우는 방식

cv2_imshow(space)

#cv2.LINE_4: 값 1 - 4방향으로 연결된 선 (상하좌우 방향).
#cv2.LINE_4: 값 4 - 4방향 연결
#cv2.LINE_8: 값 8 - 8방향 연결 (기본값)
#cv2.LINE_AA: 값 16 - 안티에일리어싱 적용
```

```python
space = np.zeros((768, 1388), dtype=np.uint8)
#여기서 dtype=np.uint8은 0~255까지

# 격자 간격 및 색상 설정
grid_spacing = 50
grid_color = 225

# 이미지에 격자 그리기
for x in range(0, space.shape[1], grid_spacing):
    cv2.line(space, (x, 0), (x, space.shape[0]), grid_color, 1)

for y in range(0, space.shape[0], grid_spacing):
    cv2.line(space, (0, y), (space.shape[1], y), grid_color, 1)
#수평선
cv2_imshow(space)

#space.shape는 이미지의 크기 (즉, 행렬의 크기)를 나타내는 튜플을 반환
print("이미지 높이:", space.shape[0])
print("이미지 너비:", space.shape[1])
```

```python
# 텐서플로는 왜쓰는가?
# 텐서(다차원 배열 ) 을 기반으로 연산 수행 하며
#1D(벡터), 2D(행렬), 3D(3차원행렬)
# 텐서플로를 사용하면 복잡한 머신 러닝 모델과 알고리즘을 쉽게 구현 가능
# 텐서플로는 데이터 플로 그래프를 사용하여 연선 표현
# 그래프 구성은 노드와, 에지로 구성 되어 있음
# 노드 - 연산 , 에지 - 노드 사이를 이동하는 데이터(텐서)
# 데이터 플로 그래프를 사용하여 텐서플로는 병렬 처리와 지연 실행을 쉽게 수행 할수 있음

# 1. 편의성-고수준 api 제공
# 2. 이식및 호환성
# 3. 확장성 - CPU,GPU 및 TPU 확장성, 분산 컴퓨팅
# 4. 유연성 -사용자 정의층, 사용자 정의 손실 함수, 새로운 최적화 전략 구현, ft.data를 이용한 데이터 파이프라인
```
